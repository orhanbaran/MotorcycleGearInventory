#Identifying Numerical and Categorical Features

```python
def grab_col_names(dataframe,cat_th=16):

    cat_cols = [col for col in df.columns if df[col].dtypes == "O"]

    num_but_cat = [col for col in df.columns if df[col].nunique() < cat_th and
                   df[col].dtypes != "O"]

    cat_cols = cat_cols + num_but_cat

    # num_cols
    num_cols = [col for col in df.columns if df[col].dtypes != "O"]

    num_cols = [col for col in num_cols if col not in num_but_cat]

    print(f"Observations: {dataframe.shape[0]}")
    print(f"Variables: {dataframe.shape[1]}")
    print(f'cat_cols: {len(cat_cols)}')
    print(f'num_cols: {len(num_cols)}')
    print(f'num_but_cat: {len(num_but_cat)}')

    return cat_cols, num_cols

cat_cols, num_cols = grab_col_names(df)
```

#Observing all the categorical features using barcharts (Top 5)

```python
def perc_on_bar(z,data = df):
    '''
    plot
    feature: categorical feature
    the function won't work if a column is passed in hue parameter
    '''

    total = len(data[z]) # Length of the column

    plt.figure(figsize = (20, 5))

    ax = sns.countplot(data = data, x = data[z], palette = 'Paired', order = data[z].value_counts().index[:5])

    for p in ax.patches:

        percentage = '{:.1f}%'.format(100 * p.get_height() / total) # Percentage of each class of the category

        x = p.get_x() + p.get_width() / 2 - 0.05 # Width of the plot

        y = p.get_y() + p.get_height()           # Hieght of the plot

        ax.annotate(percentage, (x, y), size = 12) # Annotate the percantage
        ax.set_xticklabels(ax.get_xticklabels(), rotation=45, ha='right')

    plt.show() 
# Show the plot

for columns in cat_cols:
        perc_on_bar(columns)
```

#CALLING 'BOXPLOT' FUNCTION FOR EACH CATEGORICAL VARIABLES EXCEPT 'NAME' COLUMN SINCE IT HAS TOO MANY UNIQUE VALUES

```python
#BOX PLOTS TO INTERPRET THE DATA DISTRIBUTION ON EACH VALUE WITH AND WITHOUT OUTLIERS
#FUNCTION TO DROW BOX-PLOTS
def boxplot(z):

    plt.figure(figsize = (12, 5)) # Setting size of boxplot

    sns.boxplot(x = z, y = df['price']) # Defining x and y

    plt.show()

    plt.figure(figsize = (12, 5))

    plt.title('Without Outliers')

    sns.boxplot(x = z, y = df['price'], showfliers = False) # Turning off the outliers

    plt.show()
```

```python
for cols in cat_cols:
            boxplot(df[cols])
```
